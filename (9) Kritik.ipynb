```python
#4a 

import sympy as sp

x,y = sp.symbols ('x, y')

f = sp.exp(x)*sp.sin(y)+(y**3) # function for part a
df_dx = sp.diff(f, x) # Differentiate f with respect to x
df_dy = sp.diff(f, y) # Differentiate f with respect to y

print ("partial deriv wrt x:", df_dx,"  ", "partial deriv wrt y:", df_dy)
```

```python
#4b

x0 = 1
y0 = -1
alpha = 0.01
num_iterations = 100

def grad_descent(x0, y0, f, grad_f, alpha, num_iterations):
    x = x0 #initial x value
    y = y0 #initial y value

    for i in range(num_iterations):
        grad_x, grad_y = grad_f(x, y) #gradient of f at x, y 
        x = x - (alpha*grad_x)
        y = y - (alpha*grad_y)
        return x, y 

def f(x, y):
    f = ((x**2)*y)+(x*(y**2))

def grad_f(x, y):
    grad_x = 2*x*y + y**2 
    grad_y = x**2 + 2*x*y 
    return grad_x, grad_y 

value = grad_descent(x0, y0, f, grad_f, alpha, num_iterations) 
print ("x and y are:", value) 
```

```python
#4c

import sympy as sp

x,y = sp.symbols ('x, y')

f = (((x**3)-(y**2))/((x*y)+1)) # function for part c
df_dx = sp.diff(f, x) # Differentiate f with respect to x
df_dy = sp.diff(f, y) # Differentiate f with respect to y

df_dxx = sp.diff(df_dx, x) #partial x wrt y
df_dyy = sp.diff(df_dy, y) #partial y wrt x
df_dxdy = sp.diff(df_dx, y) #mixed of x wrt y
df_dydx = sp.diff(df_dy, x) #mixed of y wrt x

print ("2nd partial of x wrt y:", df_dxx,"  ","2nd partial of y wrt x:", df_dyy,"  ","mixed partial of x wrt y:", df_dxdy,"  ","mixed partial of y wrt x:", df_dydx)
```

```python
#4d

import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

x, y = sp.symbols('x y')
j = x**3 - 3*x*y + y**3

j_func = sp.lambdify((x, y), j, 'numpy')

x_vals = np.linspace(-3, 3, 400)
y_vals = np.linspace(-3, 3, 400)
X, Y = np.meshgrid(x_vals, y_vals)
Z = j_func(X, Y)

plt.contourf(X, Y, Z, levels=50, cmap='viridis')
plt.colorbar()
plt.title('Contour plot of $j(x, y) = x^3 - 3xy + y^3$')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.show()
```

```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

x, y = sp.symbols('x y')
j = x - y

j_func = sp.lambdify((x, y), j, 'numpy')

x_vals = np.linspace(-3, 3, 400)
y_vals = np.linspace(-3, 3, 400)
X, Y = np.meshgrid(x_vals, y_vals)
Z = j_func(X, Y)

plt.contourf(X, Y, Z, levels=50, cmap='viridis')
plt.colorbar()
plt.title('Contour plot of $j(x, y) = x - y$')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.show()
```
