```python
#4a)

#defining function for part a
def func_1 (x,y):
    return x**2 + y**2

#defining partial derivative with respect to x and y for function 1
def func1_deriv(x,y):
    return 2*x, 2*y

#finding gradient descent
def grad_f_1(x,y, learning_rate):
    x0 = 0.1 #x-value for test case 1
    y0 = 0.1 #y-value for test case 1
    learning_rate: 0.1 #learning rate for test case 1

    for i in range (10): #max iteration
        grad_x, grad_y = func1_deriv (x0, y0)
        x = x0 - learning_rate * grad_x
        y = y0 - learning_rate * grad_y
    return x,y 

answer_x, answer_y = grad_f_1(0.1,0.1, 0.1)
print ("Gradient descent values are:", answer_x, answer_y)
    
```
